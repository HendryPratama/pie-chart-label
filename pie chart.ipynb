```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({'City':['Jakarta','Medan','Lampung','Palambang'],'Case':[21312,4324,5345,1243]})
data
```

```python
label = data['City'].to_list()
value = data['Case'].to_list()

fig, ax = plt.subplots(figsize=(10, 10))
patches, texts, pcts = ax.pie(value, labels=label, autopct='%.1f%%',
      textprops={'size': 'x-large'})
plt.setp(pcts, color='white', fontweight='bold')
ax.set_title('Pie Chart',fontsize=18)
plt.tight_layout()
```
